```python
import pandas as pd
import keras 
from keras.models import *
from keras.layers import *
import numpy as np
```

```python
df=pd.read_csv('air_quality_Nov2017.csv')
df
```

```python
df1=df.drop(['Generated'],1)
df1=df1.drop(['Date Time'],1)
df1.head(10)
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(axis=0)
df1.head(10)
```

```python
len(df1)
```

```python
df1.isnull().sum()
```

```python
df1['categorical'] = pd.Categorical(df1['Air Quality'])
df1.head(10)
```

```python
df_dummies = pd.get_dummies(df1['categorical'],prefix = 'quality')
df_dummies.head(10)
```

```python
df1 = pd.concat([df1, df_dummies], axis=1)
df1.head(10)
```

```python
df1 = df1.drop(['Air Quality'],1)
```

```python
df1.head(5)
```

```python
train_dataset = df1.sample(frac=0.8,random_state=100)
test_dataset=df1.drop(train_dataset.index)
```

```python
test_dataset.head(10)
len(test_dataset)
```

```python
len(train_dataset)
```

```python
targets = train_dataset[['quality_Good','quality_Moderate']]
targets.head(10)
```

```python
predictors = train_dataset.drop(train_dataset.columns[[0,3,4,6,7,9,10,12, 13, 14,]], axis=1)
predictors.head(10)
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1]
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50,activation='relu',input_shape=(n_cols,)))
    keras.layers.Dropout(0.6)
    BatchNormalization()
    model.add(Dense(20,activation='relu'))
    model.add(Dense(2))
    model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
    return model
    
```

```python
model=regression_model()
model.fit(predictors_norm, targets, validation_split=0.3, epochs=100, verbose=2,shuffle=True)
```

```python
targets_1=test_dataset[['quality_Good','quality_Moderate']]
predictors_1 = test_dataset.drop(test_dataset.columns[[0,3,4,6,7,9,10,12, 13, 14,]], axis=1)
```

```python
predictors_norm_1 = (predictors_1 - predictors_1.mean()) / predictors_1.std()
predictors_norm_1.head()
```
